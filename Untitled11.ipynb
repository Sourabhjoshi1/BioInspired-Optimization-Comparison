```python
# @title 1. Setup and Imports
import numpy as np
import time
from scipy.optimize import differential_evolution, dual_annealing
from scipy.optimize import Bounds
import matplotlib.pyplot as plt
import pandas as pd

# Increase the default number of iterations for better comparison
MAX_ITER = 100
POP_SIZE = 20

print("Required libraries imported.")
```

```python
# @title 2. Benchmark Functions

def rastrigin(X):
    """
    Rastrigin Function - Multimodal, many local minima.
    Global minimum: 0.0 at X = [0, 0, ...]
    """
    A = 10
    n = len(X)
    return A * n + sum([(x**2 - A * np.cos(2 * np.pi * x)) for x in X])

def sphere(X):
    """
    Sphere Function - Unimodal, convex.
    Global minimum: 0.0 at X = [0, 0, ...]
    """
    return sum(x**2 for x in X)

# Dictionary to easily access the functions
BENCHMARK_FUNCTIONS = {
    "Rastrigin": rastrigin,
    "Sphere": sphere
}

print("Benchmark functions defined.")
```

```python
# @title 3. Optimization Algorithm Wrappers

def optimize_de(func, bounds, max_iter, pop_size, dim):
    """Wrapper for Scipy's Differential Evolution."""
    start_time = time.time()

    # Differential Evolution (DE)
    result = differential_evolution(
        func,
        bounds,
        maxiter=max_iter,
        popsize=pop_size,
        disp=False, # Suppress intermediate output
        polish=False # No final refinement
    )

    runtime = time.time() - start_time
    return result.fun, runtime, result.nfev # function value, time, func evaluations

def optimize_sa(func, bounds, max_iter, pop_size, dim):
    """Wrapper for Scipy's Dual Annealing (a form of Simulated Annealing)."""
    # Note: Dual Annealing doesn't use a 'pop_size' directly, but we include it
    # for consistent function signature. The 'max_iter' controls the temperature steps.
    start_time = time.time()

    # Create the initial range for bounds
    # dual_annealing expects bounds as a list of (min, max) tuples
    # The Bounds object has .lb (lower bounds) and .ub (upper bounds) attributes
    bounds_list = [(bounds.lb[i], bounds.ub[i]) for i in range(len(bounds.lb))]

    result = dual_annealing(
        func,
        bounds_list,
        maxiter=max_iter,
        initial_temp=5230, # Default setting
        visit=2.62, # Default setting
        accept=-5.0 + 1e-6, # Default setting
        minimizer_kwargs=None
    )

    runtime = time.time() - start_time
    return result.fun, runtime, result.nfev # function value, time, func evaluations

# Dictionary of algorithms to compare
ALGORITHMS = {
    "Differential Evolution (DE)": optimize_de,
    "Simulated Annealing (SA)": optimize_sa
}

print("Optimization algorithm wrappers defined.")
```

```python
# @title 4. Main Comparative Analysis

# --- Experiment Parameters ---
D = 10  # Dimensionality (number of variables)
LOWER_BOUND = -5.12
UPPER_BOUND = 5.12
NUM_RUNS = 10 # Number of independent runs for statistical average

# Create Bounds object for the optimizers (shared between functions)
bounds = Bounds([LOWER_BOUND] * D, [UPPER_BOUND] * D)

results_data = []

print(f"Starting comparison (D={D}, MAX_ITER={MAX_ITER}, {NUM_RUNS} runs per test)...")

for func_name, func in BENCHMARK_FUNCTIONS.items():
    print(f"\n--- Testing Function: {func_name} ---")

    for algo_name, optimizer_func in ALGORITHMS.items():

        all_final_values = []
        all_runtimes = []
        all_nfev = []

        for run in range(NUM_RUNS):
            # Run the optimization
            final_val, runtime, nfev = optimizer_func(
                func, bounds, MAX_ITER, POP_SIZE, D
            )

            all_final_values.append(final_val)
            all_runtimes.append(runtime)
            all_nfev.append(nfev)

        # Calculate statistics
        avg_value = np.mean(all_final_values)
        std_value = np.std(all_final_values)
        avg_runtime = np.mean(all_runtimes)
        avg_nfev = np.mean(all_nfev)

        # Store the results
        results_data.append({
            'Function': func_name,
            'Algorithm': algo_name,
            'Avg. Final Value': avg_value,
            'Std. Final Value': std_value,
            'Avg. Runtime (s)': avg_runtime,
            'Avg. NFEV': avg_nfev
        })

        print(f"  {algo_name:<30} | Avg. Val: {avg_value:.4f} (Std: {std_value:.4f}) | Avg. Time: {avg_runtime:.4f}s")

# Convert results to a DataFrame for easy viewing
df_results = pd.DataFrame(results_data)
print("\nComparison Complete!")
```

```python
# @title 5. Visualization and Results

## Results Table
print("\n## Comparative Analysis Results Table")
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 1000)
pd.set_option('display.colheader_justify', 'center')
print(df_results)

print("\n---")
print("Avg. Final Value (Lower is Better) and Avg. Runtime (s) are the key metrics.")

# --- Visualization ---
plt.style.use('seaborn-v0_8-whitegrid')

# 1. Performance (Final Value) Plot
fig, ax = plt.subplots(figsize=(10, 6))
df_results.pivot_table(index="Algorithm", columns="Function", values="Avg. Final Value").plot(kind='bar', ax=ax)
ax.set_title(f'Avg. Final Objective Value (D={D}, Runs={NUM_RUNS})', fontsize=14)
ax.set_ylabel('Average Final Value (log scale)', fontsize=12)
ax.set_xlabel('Algorithm', fontsize=12)
ax.set_yscale('log') # Use log scale for better visualization of small differences
plt.xticks(rotation=0)
plt.legend(title='Function', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

# 2. Runtime Plot
fig, ax = plt.subplots(figsize=(10, 6))
df_results.pivot_table(index="Algorithm", columns="Function", values="Avg. Runtime (s)").plot(kind='bar', ax=ax, colormap='Pastel1')
ax.set_title(f'Avg. Runtime (s) (D={D}, Runs={NUM_RUNS})', fontsize=14)
ax.set_ylabel('Average Runtime (s)', fontsize=12)
ax.set_xlabel('Algorithm', fontsize=12)
plt.xticks(rotation=0)
plt.legend(title='Function', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print("\n--- Analysis Summary ---")
print("1. **Rastrigin Function (Multimodal)**: This function tests the algorithm's global search capability. The algorithm that achieves the lowest average final value is generally considered superior in exploration.")
print("2. **Sphere Function (Unimodal)**: This function tests the algorithm's convergence speed and exploitation capability. The algorithm with the lowest average final value and shortest runtime, while maintaining low NFEV, is typically preferred.")
print("3. **Std. Final Value**: A low standard deviation indicates a more *robust* algorithm that consistently finds a similar quality solution across multiple runs.")
```
